```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
IMG_SIZE = 64
SEQ_LENGTH = 20

classes = [
    "Basketball",
    "Biking",
    "Diving",
    "HorseRace",
    "TennisSwing",
    "Fencing",
    "GolfSwing",
    "Lunges"
]
```

```python
def extract_frames(video_path):
    
    frames = []
    video = cv2.VideoCapture(video_path)
    
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    skip = max(total_frames // SEQ_LENGTH, 1)
    
    for i in range(SEQ_LENGTH):
        video.set(cv2.CAP_PROP_POS_FRAMES, i * skip)
        success, frame = video.read()
        
        if not success:
            break
        
        frame = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
        frame = frame / 255.0
        
        frames.append(frame)
    
    video.release()
    
    return np.array(frames)
```

```python
model_path = "D:\last term\deep learning\CNN-LSTM\HAR_CNN_LSTM_model.h5"

model = load_model(model_path)

print("Model Loaded Successfully!")
```

```python
video_path = r"31.03.2026_21.52.15_REC.mp4"

frames = extract_frames(video_path)
frames = np.expand_dims(frames, axis=0)

prediction = model.predict(frames)

predicted_class = classes[np.argmax(prediction)]

print("Predicted Activity =", predicted_class)


import cv2

cap = cv2.VideoCapture(video_path)

output_path = "prediction_output.mp4"

frame_width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))


while True:
    
    ret, frame = cap.read()
    
    if not ret:
        break
    
    cv2.putText(frame,
                f"Activity: {predicted_class}",
                (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (0,255,0),
                2,
                cv2.LINE_AA)
    
    out.write(frame)
    
    cv2.imshow("HAR Prediction", frame)
    
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break


cap.release()
out.release()   
cv2.destroyAllWindows()
```
